```python
qualified = [item for item in nlp.vocab if item.has_vector and item.is_alpha]
```

```python
lexmap = []
t = annoy.AnnoyIndex(300)
for i, item in enumerate(islice(sorted(qualified, key=lambda x: x.prob, reverse=True), 100000)):
    t.add_item(i, item.vector)
    lexmap.append(item)
t.build(25)
```

```python
def similarsemantic(t, nlp, word, n):
    seen = set()
    count = 0
    for i in t.get_nns_by_vector(nlp.vocab[word].vector, 100):
        this_word = lexmap[i].orth_.lower()
        if this_word not in seen and word != this_word:
            seen.add(this_word)
            count += 1
            yield this_word
            if count >= n:
                break
list(similarsemantic(t, nlp, "die", 20))
```

```python
##Macbeth sleep no more

text = """\

thought I heard a voice cry, “Sleep no more! Your are murdering sleep.” Innocent sleep. 
Sleep that soothes away all our worries. Sleep that puts each day to rest. 
Sleep that relieves the weary laborer and heals hurt minds. Sleep, the main course in 
life’s feast, and the most nourishing.

What are you talking about?

The voice kept crying, “Sleep no more!” to everyone in the house. “You has murdered sleep, 
and therefore you will sleep no more.”

Who said that? Why, you let yourself become weak when you think about things in this cowardly way. 

I can’t go back. I’m afraid even to think about what I’ve done. I can’t stand to look at it again.

Coward! Give me the daggers. Dead and sleeping people can’t hurt you any more than pictures can. 
Only children are afraid of scary pictures. 

"""
```

```python
##minor change in words, wierd context
chance = 30
nuttiness = 30
doc = nlp(text)
output = []
for tok in doc:
    if tok.is_alpha and random.randrange(100) < chance:
        output.append(
            random.choice(list(similarsemantic(t, nlp, tok.text, nuttiness))) + tok.whitespace_)
    else:
        output.append(tok.text_with_ws)
print("".join(output))
```

```python
def meanv(vecs):
    total = np.sum(vecs, axis=0)
    return total / len(vecs)
```
